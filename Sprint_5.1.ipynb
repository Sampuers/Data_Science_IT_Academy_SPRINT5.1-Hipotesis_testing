###### Amb el mateix dataset selecciona dos altres atributs del conjunt de dades. Calcula els p-valors i digues si rebutgen la hipòtesi nul·la agafant un alfa de 5%.

Elijo la variable de 'Rk'y 'W', que representa posición del equipo dentro de la liga y partidos ganados.

PASO 1, Establezco una hipótesis nula y una alternativa

- H0: las medias de las muestras son iguales.
- H1: las medias de las muestras son desiguales.

- Alpha: 5%

PASO 2: Visualizo las variables y los datos son numéricos. 

```python
columnas = df[['Rk', 'W']]
columnas
```

PASO 3, realizo la prueba estadística T de Student para probar si las muestras son significativamente diferentes. 

```python
from scipy.stats import ttest_ind
# Establezco las dos muestras
muestra1 = df['Rk']
muestra2 = df['W']
# Realizo la prueba t Student

estadistico, p_value = ttest_ind(muestra1, muestra2)
print('Estadistico=%.3f, p_value=%.3f' % (estadistico, p_value))

```

PASO 4, En este caso, el pvalue, indica que está por debajo de 0,05 y por lo tanto sí podemos rechazar la hipótesis nula y dar por confirmada que las medias de las muestras son desiguales.

### - Exercici 3
###### Continua amb el conjunt de dades adjunt i selecciona tres atributs del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%.

Elijo la variable de 'Rk', la variable de 'W' y la variable de 'GF'  que representan, ranking de liga, partidos ganados y goles a favor.

PASO 1, Establezco una hipótesis nula y una alternativa

- H0: las medias de las muestras son iguales.
- H1: una o más de las medias de las muestras son desiguales.
- Alpha: 5%

PASO 2: Visualizo las variables y los datos son numéricos. 

```python
columnas3 = df[['Rk', 'W', 'GF']]
columnas3
```

PASO 3, realizo la prueba estadística ANOVA para probar si las muestras son significativamente diferentes.

```python
from scipy.stats import f_oneway
# Establezco las dos muestras
variable1 = df['Rk']
variable2 = df['W']
variable3 = df['GF']
# Realizo la prueba ANOVA

estadistico, p_value = f_oneway(variable1,variable2,variable3)
print('Estadistico=%.3f, p_value=%.3f' % (estadistico, p_value))
```

PASO 4, En este caso, el pvalue, indica que está por debajo de 0,05 y por lo tanto sí podemos rechazar la hipótesis nula y dar por confirmada que una o más de las medias de las muestras no son iguales.
